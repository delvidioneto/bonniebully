```python
from bonniebully import intdate
from datetime import date, datetime
import pandas
```

```python
dia_atual = date(day=13, month=11, year=2026)
print(dia_atual)
```

```python
start = intdate('MONTH', dia_atual, 0, 'B', CalendarType="FISCAL").getDates()
end = intdate('MONTH', dia_atual, 0, 'E', CalendarType="FISCAL").getDates()
print(start, end)
```

```python

dia_util = intdate('BDAY', dia_atual, 3, 'S', "BR", "SP").getDates()
print(dia_util)
```
